#### Parametro batzuk kalkulatzen
Behin 25 taldeak ditugula, parametro batzuk aterako ditugu datuetatik, pisu-matrize egokia egiteko.

```python
# Home/Away multipliers

# Function to determine the result based on scores
def determine_result(row):
    """Determines the result based on home and away scores."""
    if row['home_score'] > row['away_score']:
        return "Home Win"
    elif row['home_score'] < row['away_score']:
        return "Away Win"
    else:
        return "Draw"

# Apply the function to create a 'result' column
filtered_df1 = filtered_df1.copy()  # Create an explicit copy
filtered_df1.loc[:, 'result'] = filtered_df1.apply(determine_result, axis=1)

# Calculate the counts
home_wins = filtered_df1['result'].value_counts().get("Home Win", 0)  # Get count of "Home Win", default to 0 if not found
away_wins = filtered_df1['result'].value_counts().get("Away Win", 0)  # Get count of "Away Win", default to 0 if not found
draws = filtered_df1['result'].value_counts().get("Draw", 0)  # Get count of "Draw", default to 0 if not found

# Print the results
print(f"Home Wins: {home_wins}")
print(f"Away Wins: {away_wins}")
print(f"Draws: {draws}")

# Home/Away Multipliers
# Smoothing constant (adjust as needed)
smoothing_constant = draws+home_wins+away_wins
# Multipliers with smoothing
home_multi = round((away_wins + smoothing_constant) / (home_wins + smoothing_constant), 3)
away_multi = round(1 / home_multi, 3)

print("Home multiplier: " + str(home_multi))
print("Away multiplier: " + str(away_multi))
```

```python
# Goals puntuation
# Calculate total goals scored by home team
avg_home_goals = round(filtered_df1['home_score'].sum()/len(filtered_df1), 3)
# Calculate total goals scored by away team
avg_away_goals = round(filtered_df1['away_score'].sum()/len(filtered_df1), 3)
# Calculate average goals per match
avg_goals = round(avg_home_goals + avg_away_goals, 3)
# Print the results
print("Average Home Goals:", avg_home_goals)
print("Average Away Goals:", avg_away_goals)
print("Average Goals per Match:", avg_goals)

def goals_puntuation(goals):
  if (goals>=1):
    return (np.log2(goals) + 1) / avg_goals
  else:
    return 0
```

```python
# Tournament multipliers
# Get unique tournament names
tournaments = filtered_df1['tournament'].unique()
print("Number of tournaments: " + str(len(tournaments)))

def get_tournament_multiplier(tournament):
    """Returns a multiplier based on the tournament tier."""
    if tournament in tier1_tournaments:
        return 1.5
    elif tournament in tier2_tournaments:
        return 1.25
    elif tournament in tier3_tournaments:
        return 1
    elif tournament in tier5_tournaments:
        return 0.5
    else:
        return 0.75  # Default to tier 4 (or any other value you deem appropriate)
```

#### Pisu-matrizea eraikitzen
Orain, atera ditgun parametroekin prest gaude gure LOP problemako matrizea sortzeko, partidu bakoitzaren datuak erabiliz.

```python
# Create an empty LOP matrix
lop_matrix = pd.DataFrame(0.0, index=filtered_teams, columns=filtered_teams, dtype=float)

filtered_matches = filtered_df1[filtered_df1['home_team'].isin(filtered_teams) & filtered_df1['away_team'].isin(filtered_teams)]

# Populate LOP matrix
for index, row in filtered_matches.iterrows():
    home_team = row['home_team']
    away_team = row['away_team']

    # Skip if home_team and away_team are the same (diagonal element)
    if home_team == away_team:
        continue  # Go to the next iteration

    home_score = row['home_score']
    away_score = row['away_score']

    home_away = lop_matrix.loc[home_team, away_team]
    away_home = lop_matrix.loc[away_team, home_team]


    # Applying win/draw/loss points based on result and neutral status
    if row['result'] == "Home Win":
        if row['neutral'] == "TRUE":
            home_away += 3
        else:
            home_away += 3 * home_multi
    elif row['result'] == "Away Win":
        if row['neutral'] == "TRUE":
            away_home += 3
        else:
            away_home += 3 * away_multi
    else:  # Draw
        home_away += 1
        away_home += 1
        # Check for shootouts in df2 and award extra point to the winner
        shootout_match = df2[(df2['home_team'] == row['home_team']) &
                            (df2['away_team'] == row['away_team']) &
                            (df2['date'] == row['date'])]

        if not shootout_match.empty:  # Check if shootout occurred
            winner = shootout_match['winner'].iloc[0]  # Get the winner from df2
            if winner == row['home_team']:
                home_away += 1
            elif winner == row['away_team']:
                away_home += 1

        if row['neutral'] == "FALSE":  # Apply multipliers for non-neutral draws
            home_away *= home_multi
            away_home *= away_multi


    # Adding goal points using goals_puntuation function
    home_away += goals_puntuation(home_score)
    away_home += goals_puntuation(away_score)

    # Applying tournament multiplier
    home_away *= get_tournament_multiplier(row['tournament'])
    away_home *= get_tournament_multiplier(row['tournament'])

    # Applying date multiplier
    date_string = row['date']
    date_object = datetime.strptime(date_string, "%Y-%m-%d")  # Parse the string into a datetime object
    year = date_object.year
    home_away *= year/2021
    away_home *= year/2021

    # Smoothing
    home_away /= (np.log(filtered_teams_matches_counts[home_team]) + 1)
    away_home /= (np.log(filtered_teams_matches_counts[away_team]) + 1)

    lop_matrix.loc[home_team, away_team] = home_away
    lop_matrix.loc[away_team, home_team] = away_home

# Display the LOP matrix
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
sns.heatmap(lop_matrix, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title("LOP Matrix Heatmap")
plt.xlabel("Away Team")
plt.ylabel("Home Team")
plt.tight_layout()
plt.show()
```

#### Helburu-funtzioa

LOP problema baten helburu-funtzioa da, baina eraldaketa bat gehitu diogu, alpha balio bat sartu dugu parametro gisa ranking-eko aurreko posizioak atzekoak baino garrantzi gehiago izateko.

```python
def lop_objective_function(solution, lop_matrix, alpha=3):
    objective_value = 0
    num_teams = len(solution)
    for i in range(num_teams):
        for j in range(i + 1, num_teams):
            team1_index = solution[i]
            team2_index = solution[j]

            # The following lines have been modified to use .iloc to access data by position
            # rather than by label. This assumes that 'solution' contains numerical indices.
            team1 = lop_matrix.index[lop_matrix.index.get_loc(team1_index)]  # Get index label using get_loc
            team2 = lop_matrix.index[lop_matrix.index.get_loc(team2_index)]  # Get index label using get_loc

            # Calculate weight based on ranking position
            weight = (num_teams - i)**alpha/25  # Higher weight for teams near the top

            # Add contribution if ranking agrees with the matrix
            if lop_matrix.loc[team1, team2] > lop_matrix.loc[team2, team1]:
                objective_value += lop_matrix.loc[team1, team2] * weight

    return objective_value
```

```python
# Number of random solutions to generate
num_solutions = 100

# Generate random solutions, calculate objective values, and store them
solutions_and_values = []
for _ in range(num_solutions):
    random_solution = filtered_teams.copy()
    random.shuffle(random_solution)
    objective_value = lop_objective_function(random_solution, lop_matrix)
    solutions_and_values.append((random_solution, objective_value))

# Plot the objective values against the solutions
solution_labels = [str(i + 1) for i in range(num_solutions)]  # Solution labels for x-axis
objective_values = [value for _, value in solutions_and_values]

plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(solution_labels, objective_values)
plt.xlabel("Solutions")
plt.ylabel("Objective Function Value")
plt.title("Objective Values for Random Solutions")
plt.xticks([])  # Remove x-axis tick labels
plt.tight_layout()
plt.show()
```

```python
# Find the best and worst solutions
best_solution, best_value = max(solutions_and_values, key=lambda item: item[1])
worst_solution, worst_value = min(solutions_and_values, key=lambda item: item[1])

# Print the best and worst solutions
print("Best Solution:", best_solution, "\nObjective Value:", best_value)
print("\nWorst Solution:", worst_solution, "\nObjective Value:", worst_value)
```

###Random Search

```python
import random

def random_search_ifr(lop_matrix, num_iterations=100000):
    """
    Implementación del Random Search para el problema de ranking de fútbol.

    Args:
        lop_matrix: La matriz LOP que representa las relaciones entre los equipos.
        num_iterations: El número de iteraciones (evaluaciones de soluciones) a realizar.

    Returns:
        La mejor solución encontrada (un ranking de equipos) y su valor de fitness.
    """

    num_teams = len(lop_matrix.index)  # Número de equipos
    teams = lop_matrix.index.tolist()  # Lista de equipos

    best_solution = None  # Inicializar la mejor solución
    best_fitness = -float('inf')  # Inicializar el mejor fitness con un valor muy bajo

    for _ in range(num_iterations):
        # Generar una solución aleatoria (un ranking de equipos)
        solution = random.sample(teams, num_teams)

        # Evaluar la solución utilizando la función objetivo
        fitness = lop_objective_function(solution, lop_matrix)

        # Actualizar la mejor solución si se encuentra una mejor
        if fitness > best_fitness:
            best_fitness = fitness
            best_solution = solution.copy()  # Copiar la solución para evitar modificaciones

    return best_solution, best_fitness
```

### Algoritmo eraikitzailea
Taldeen ordena erabakitzeko, talde bakoitzaren aldeko puntuen eta talde horren aurkako puntuen diferentzia kalkulatzen da eta diferentzia positiboenaren arabera taldeak soluziora gehitzen dira. Diferentzia hori kalkulatzeko talde bakoitzaren errenkadaren batuketaren eta bere zutabearen batuketaren arteko kenketa egiten da.
Bestalde, `lop_matrix`-aren erro karratua egiten dugu, balio moderatuen alde egiteko eta muturrekoak penalizatzeko, errendimendu erregularragoa duen taldeari lehentasuna emanez.


```python
def greedy_ifr(lop_matrix):

    # Get the NumPy array from the DataFrame
    lop_matrix_np = np.sqrt(lop_matrix.to_numpy())
    n = lop_matrix_np.shape[0]  # Number of teams
    ranking = []  # Initialize empty ranking
    row_sums = lop_matrix_np.sum(axis=1)  # Calculate initial row sums
    col_sums = lop_matrix_np.sum(axis=0)  # Calculate initial col sums
    diff = row_sums - col_sums    # Calculate the difference

    while len(ranking) < n:
      max_row_col = np.argmax(diff)
      team_name = lop_matrix.index[max_row_col]
      ranking.append(team_name)
      diff[max_row_col] = -np.inf

    return ranking
```

### Bilaketa lokala
Simulateed Anneling algoritmo inplementatuko dugu.

```python
def simulated_annealing_ifr(lop_matrix, initial_temp=1000, cooling_rate=0.95, num_iterations=1000):

    # get all the teams from the index of the lop_matrix
    current_solution = list(lop_matrix.index)
    # Initialize the current solution
    random.shuffle(current_solution) # Shuffle to obtain random initial solution
    current_value = lop_objective_function(current_solution, lop_matrix)
    best_solution = current_solution.copy()
    best_value = current_value
    temperature = initial_temp

    # Lists to store values for plotting
    fitness_values = []

    for _ in range(num_iterations):
        # Generate a neighbor solution
        neighbor = current_solution.copy()
        i, j = random.sample(range(len(neighbor)), 2)  # Select two random indices
        neighbor[i], neighbor[j] = neighbor[j], neighbor[i]  # Swap elements

        # Calculate the objective value for the neighbor
        neighbor_value = lop_objective_function(neighbor, lop_matrix)

        # Decide whether to accept the neighbor
        delta = neighbor_value - current_value

        # Check if delta / temperature might cause overflow
        try:
            acceptance_prob = np.exp(delta / temperature)
        except OverflowError:
            acceptance_prob = float('inf')  # Set to infinity if overflow

        if delta > 0 or random.random() < acceptance_prob:
            current_solution = neighbor.copy()
            current_value = neighbor_value

            # Update the best solution if necessary
            if current_value > best_value:
                best_solution = current_solution.copy()
                best_value = current_value

        # Store values for plotting
        fitness_values.append(best_value)

        # Cool down the temperature
        temperature *= cooling_rate

    return best_solution, fitness_values
```

### Algoritmo ebolutiboa
EDA algoritmoaren hobekuntza bat inplementatuko dugu.

```python
def eda_ifr(lop_matrix, population_size=100, num_generations=100, selection_rate=0.5, mutation_rate=0.1):

    num_teams = len(lop_matrix.index)
    teams = lop_matrix.index.tolist()
    population = [random.sample(teams, num_teams) for _ in range(population_size)]
    best_fitness_values = []

    for generation in range(num_generations):
        fitness_values = [lop_objective_function(solution, lop_matrix) for solution in population]
        selected_indices = np.argsort(fitness_values)[-int(selection_rate * population_size):]
        selected_population = [population[i] for i in selected_indices]

        # Probability model based on team positions:
        probability_model = build_position_based_model(selected_population)

        new_population = []
        for _ in range(population_size):
            new_solution = sample_from_position_based_model(probability_model, teams)

            # Apply mutation:
            if random.random() < mutation_rate:
                new_solution = mutate_solution(new_solution)

            new_population.append(new_solution)

        population = new_population

        best_solution_current_gen = max(population, key=lambda solution: lop_objective_function(solution, lop_matrix))
        best_fitness_current_gen = lop_objective_function(best_solution_current_gen, lop_matrix)
        best_fitness_values.append(best_fitness_current_gen)

    best_solution_overall = max(population, key=lambda solution: lop_objective_function(solution, lop_matrix))
    return best_solution_overall, best_fitness_values


def build_position_based_model(selected_population):
    """Builds a probability model based on team positions."""
    num_teams = len(selected_population[0])
    probability_model = {i: {} for i in range(num_teams)} # Dictionary to store probabilities

    for solution in selected_population:
        for position, team in enumerate(solution):
            probability_model[position][team] = probability_model[position].get(team, 0) + 1

    # Normalize probabilities:
    for position in probability_model:
        total_count = sum(probability_model[position].values())
        for team in probability_model[position]:
            probability_model[position][team] /= total_count

    return probability_model


def sample_from_position_based_model(probability_model, teams):
    """Samples a new solution based on position probabilities."""
    num_teams = len(teams)
    solution = [None] * num_teams
    available_teams = teams[:]

    for position in range(num_teams):
        # Get probabilities for this position
        position_probs = probability_model[position]

        # Filter probabilities for available teams only
        available_probs = {team: prob for team, prob in position_probs.items() if team in available_teams}

        # If no probabilities for available teams (can happen due to randomness), choose randomly
        if not available_probs:
            chosen_team = random.choice(available_teams)
        else:
            # Normalize probabilities for available teams
            total_prob = sum(available_probs.values())
            probs = {team: prob / total_prob for team, prob in available_probs.items()}

            # Choose a team based on probabilities
            chosen_team = random.choices(list(probs.keys()), weights=list(probs.values()), k=1)[0]

        # Add chosen team to solution and remove from available teams
        solution[position] = chosen_team
        available_teams.remove(chosen_team)

    return solution

def mutate_solution(solution):
    """Applies a swap mutation to the solution."""
    # Select two random indices
    index1, index2 = random.sample(range(len(solution)), 2)

    # Swap the elements at the selected indices
    solution[index1], solution[index2] = solution[index2], solution[index1]

    return solution
```

```python
def instance_generator(df, instance_size, num_instances):

    # Empty list to store the instances
    instances = []

    for i in range(num_instances):
        # we use np.random.choice with replace=False to ensure unique indices
        random_indices = np.random.choice(num_rows, size=num_rows_to_select, replace=False)

        # Create the new instance (DataFrame) with the selected rows
        new_instance = filtered_df1.iloc[random_indices]

        # Filter the DataFrame to include matches from tier1 tournaments:
        new_instance_matches = new_instance[new_instance['tournament'].isin(tier1_tournaments)]

        # Calculate the number of matches played by each team in new_instance_matches
        new_instance_teams_matches_counts = new_instance_matches['home_team'].value_counts() + new_instance_matches['away_team'].value_counts()

        # Get the teams that have played the most matches in the World Cup:
        new_instance_teams = new_instance_teams_matches_counts.nlargest(25).index.tolist()



        # Home/Away multipliers

        # Apply the function to create a 'result' column
        new_instance = new_instance.copy()  # Create an explicit copy
        new_instance.loc[:, 'result'] = new_instance.apply(determine_result, axis=1)

        # Calculate the counts
        home_wins = new_instance['result'].value_counts().get("Home Win", 0)  # Get count of "Home Win", default to 0 if not found
        away_wins = new_instance['result'].value_counts().get("Away Win", 0)  # Get count of "Away Win", default to 0 if not found
        draws = new_instance['result'].value_counts().get("Draw", 0)  # Get count of "Draw", default to 0 if not found

        # Home/Away Multipliers
        # Smoothing constant (adjust as needed)
        smoothing_constant = draws+home_wins+away_wins
        # Multipliers with smoothing
        home_multi = round((away_wins + smoothing_constant) / (home_wins + smoothing_constant), 3)
        away_multi = round(1 / home_multi, 3)

        # Goals puntuation
        # Calculate total goals scored by home team
        avg_home_goals = round(new_instance['home_score'].sum()/len(new_instance), 3)
        # Calculate total goals scored by away team
        avg_away_goals = round(new_instance['away_score'].sum()/len(new_instance), 3)
        # Calculate average goals per match
        avg_goals = round(avg_home_goals + avg_away_goals, 3)

        # Tournament multipliers
        # Get unique tournament names
        tournaments = new_instance['tournament'].unique()



        # Create an empty LOP matrix
        new_instance_lop_matrix = pd.DataFrame(0.0, index=new_instance_teams, columns=new_instance_teams, dtype=float)

        new_instance_matches = new_instance[new_instance['home_team'].isin(new_instance_teams) & new_instance['away_team'].isin(new_instance_teams)]

        # Populate LOP matrix
        for index, row in new_instance_matches.iterrows():
            home_team = row['home_team']
            away_team = row['away_team']

            # Skip if home_team and away_team are the same (diagonal element)
            if home_team == away_team:
                continue  # Go to the next iteration

            home_score = row['home_score']
            away_score = row['away_score']

            home_away = new_instance_lop_matrix.loc[home_team, away_team]
            away_home = new_instance_lop_matrix.loc[away_team, home_team]


            # Applying win/draw/loss points based on result and neutral status
            if row['result'] == "Home Win":
                if row['neutral'] == "TRUE":
                    home_away += 3
                else:
                    home_away += 3 * home_multi
            elif row['result'] == "Away Win":
                if row['neutral'] == "TRUE":
                    away_home += 3
                else:
                    away_home += 3 * away_multi
            else:  # Draw
                home_away += 1
                away_home += 1
                # Check for shootouts in df2 and award extra point to the winner
                shootout_match = df2[(df2['home_team'] == row['home_team']) &
                                    (df2['away_team'] == row['away_team']) &
                                    (df2['date'] == row['date'])]

                if not shootout_match.empty:  # Check if shootout occurred
                    winner = shootout_match['winner'].iloc[0]  # Get the winner from df2
                    if winner == row['home_team']:
                        home_away += 1
                    elif winner == row['away_team']:
                        away_home += 1

                if row['neutral'] == "FALSE":  # Apply multipliers for non-neutral draws
                    home_away *= home_multi
                    away_home *= away_multi


            # Adding goal points using goals_puntuation function
            home_away += goals_puntuation(home_score)
            away_home += goals_puntuation(away_score)

            # Applying tournament multiplier
            home_away *= get_tournament_multiplier(row['tournament'])
            away_home *= get_tournament_multiplier(row['tournament'])

            # Applying date multiplier
            date_string = row['date']
            date_object = datetime.strptime(date_string, "%Y-%m-%d")  # Parse the string into a datetime object
            year = date_object.year
            home_away *= year/2021
            away_home *= year/2021

            # Smoothing
            home_away /= (np.log(new_instance_teams_matches_counts[home_team]) + 1)
            away_home /= (np.log(new_instance_teams_matches_counts[away_team]) + 1)

            new_instance_lop_matrix.loc[home_team, away_team] = home_away
            new_instance_lop_matrix.loc[away_team, home_team] = away_home

        # Store the new instance
        instances.append(new_instance_lop_matrix)

    return instances
```

## Similarity (laguntzailea)
Algoritmo bakoitzarekin lortutako soluzioen antzekotasuna ikusteko funtzioa.

```python
def similarity(solution1, solution2):
    """
    Calcula la semejanza entre dos soluciones utilizando la distancia de Levenshtein.

    Args:
        solution1: La primera solución (lista de elementos).
        solution2: La segunda solución (lista de elementos).

    Returns:
        Un valor entre 0 y 1 que representa la semejanza entre las dos soluciones.
    """

    # Asegúrate de que ambas soluciones tengan la misma longitud
    if len(solution1) != len(solution2):
        raise ValueError("Las soluciones deben tener la misma longitud.")

    # Calcula la distancia de Levenshtein entre las dos soluciones
    distance = levenshtein_distance(solution1, solution2)

    # Normaliza la distancia para obtener un valor de semejanza entre 0 y 1
    max_distance = len(solution1)
    similarity = 1 - (distance / max_distance)

    return similarity


def levenshtein_distance(s1, s2):
    """
    Calcula la distancia de Levenshtein entre dos cadenas o listas.
    """

    if len(s1) < len(s2):
        return levenshtein_distance(s2, s1)

    if len(s2) == 0:
        return len(s1)

    previous_row = range(len(s2) + 1)
    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row

    return previous_row[-1]
```
